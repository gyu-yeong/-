## 네이버 로그인 자동화

```python
from selenium import webdriver
browser = webdriver.Chrome('C:/Users/user/chromedriver.exe') #크롬드라이브 경로

#1. 네이버로 이동
browser.get('http://naver.com')

#2. 네이버 로그인 클릭
elem = browser.find_elements_class_name('link_login') #네임이 link_login인 것
elem.click()

#3. 아이디 비번 입력
browser.find_elements_id('id').send_keys('naver_id') #id가 id인 것
browser.find_elements_id('password').send_keys('password') #id가 password인 것
#send_keys('naver_id') -> 괄호 안에는 내 비번과 아이디 입력 

#4. 로그인 버튼 클릭 
browser.find_elements_id('log.login').click() #id가 log.login인 것
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

#1. 구글드라이브 실행 및 URL 접속
browser = webdriver.Chrome('C:/Users/user/chromedriver.exe') #크롬드라이브 경로
browser.get("https://www.naver.com")

elem = browser.find_element(By.CLASS_NAME, 'link_login')
elem.click()

#3. 아이디 비번 입력
browser.find_element(By.ID,'id').send_keys('naver_id') #id가 id인 것
browser.find_element(By.ID,'pw').send_keys('password')
#send_keys('naver_id') -> 괄호 안에는 내 비번과 아이디 입력 

#4. 로그인 버튼 클릭 
browser.find_element(By.ID,'log.login').click() #id가 log.login인 것
time.sleep(3)

#5. 새로운 id 입력
browser.find_element(By.ID,'id').clear() #이전에 있던 아이디 clear해줘야 함 

browser.find_element(By.ID,'id').send_keys('koaroo1202')

#6. html 정보 출력
print(browser.page_source) #지금 페이지에 있는 모든 html 문서 출력

#7. 브라우저 종료
# browser.close(0) #현재 탭만 종료
browser.quit() #전체 브라우저 종료
```

### 방법2: 민지가 준 코드 돌리기
#### 코드 하나씩 돌릴때마다 창 닫지말고 냅두기

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

from selenium.webdriver.support.ui import WebDriverWait #특정 elem이 나올때까지 기다리는 
from selenium.webdriver.support import expected_conditions as EC
from datetime import date
```

```python
#1. 구글드라이브 실행 및 URL 접속
browser = webdriver.Chrome('C:/Users/user/chromedriver.exe') #크롬드라이브 경로
browser.maximize_window() # 창 최대화 

url = 'https://flight.naver.com/' #사이트 정보
browser.get(url) #url로 이동 
```

```python
def wait_until(xpath_str):
    WebDriverWait(browser,30).until(EC.presence_of_element_located
    ((By.XPATH,xpath_str))) #By.XPATH,xpath_str이게 나올때까지 30초만 기다려라는 def문
```

```python
#가는 날 클릭
browser.find_element(By.XPATH,'//*[@id="__next"]/div/div[1]/div[4]/div/div/div[2]/div[2]/button[1]').click()
```

```python
# 이번 달  27, 28일 선택

##현재월 / 설정하고 싶은 월 
now=date.today().month
count=11-now

#출발일 27
wait_until('//b[text()="27"]') #text로 27이 나올때까지 기다리라는 의미 
day27=browser.find_elements(By.XPATH, '//b[text()="27"]')
day27[count].click()

#출발일 28
wait_until('//b[text()="28"]')
day28=browser.find_elements(By.XPATH, '//b[text()="28"]')
day28[count].click()

#도착
wait_until('//b[text()="도착"]')
arrival=browser.find_element(By.XPATH, '//b[text()="도착"]')
arrival.click()

wait_until('//button[text()="유럽"]')
europe=browser.find_element(By.XPATH, '//button[text()="유럽"]')
europe.click()
```
