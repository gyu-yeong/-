### 줄바꿈을 위해 의미없는 'tr'들이 있음
### 이렇게 의미없는 'tr'의 안을 들여다보면 'td'가 하나밖에 없다는 특징이 있음
### 따라서 이를 반복문 안에 활용

```python
url = "https://finance.naver.com/sise/sise_market_sum.nhn?sosok=0&page="

for i in range(1,2):
    res = requests.get(url+str(i))
    res.raise_for_status()
    soup = BeautifulSoup(res.text, 'lxml')
    data_rows = soup.find('table',attrs={'class':'type_2'}).find('tbody').find_all('tr')
    for row in data_rows:
        columns = row.find_all('td')
        if len(columns) <= 1: #td가 1 이하인 데이터는 스킵
            continue
        data = [column.get_text().strip() for column in columns] #strip()를 이용해 \t같은 공백 제거
        print(data)
```

### csv에 저장하기까지 풀코드

```python
import csv
import requests
from bs4 import BeautifulSoup

filename = '시가총액1-200.csv'
f = open(filename,'w',encoding='utf-8-sig', newline='') #newline은 한 회사의 정보가 다 출력되면 엔터가 되어야 하니까 
#'w'는 쓰기모드                                                #utf-8-sig는 엑셀 파일에서 열 때 한글깨지면 이걸루
   
writer = csv.writer(f)     
title = 'N 종목명 현재가 전일비 등락률 액면가 시가총액 상장주식수 외국인비율 거래량 PER ROE 토론실'.split('\t') 
#띄어쓰기 기준으로 스플릿
print(type(title))
writer.writerow(title) #이걸 했는데 왜 컬럼명이 생겼지?  -> writerow가 행 기준으로 먼저 넣어줌 for문보다 앞에 있어서 1행에 옴

url = "https://finance.naver.com/sise/sise_market_sum.nhn?sosok=0&page="

for i in range(1,5):
    res = requests.get(url+str(i))
    res.raise_for_status()
    soup = BeautifulSoup(res.text, 'lxml')
    data_rows = soup.find('table',attrs={'class':'type_2'}).find('tbody').find_all('tr')
    for row in data_rows:
        columns = row.find_all('td')
        if len(columns) <= 1: #td가 1 이하인 데이터는 스킵
            continue
        data = [column.get_text().strip() for column in columns] #strip()를 이용해 \t같은 공백 제거
        writer.writerow(data) #list 형태로 넣어줌
```
