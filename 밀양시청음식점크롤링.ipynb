## 구현 코드

```python
#headers에 user agent 정보 넣기
headers = {'User-Agent' : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"}

#리스트 생성
noun=[]
category=[]
menu=[]
which=[]


#주소 url sjgrl
for i in range(0, 28):
    url =f"https://www.miryang.go.kr/tur/EgovRestaurantList.do?mnNo=50100000000&searchKeywordTo=&searchCondition=&searchKeyword=&pageIndex={i}"
    
    #request.get()으로 연결
    r =requests.get(url, headers=headers)
    r.status_code
    
    #beautifulsoup로 접속하기
    soup=BeautifulSoup(r.text, 'html.parser') #text 또는 content
    
    #정보가 담긴 클래스나 태그 찾기
    info = soup.find_all('ul', class_=['Tu-uL'])
    name = soup.find_all('h3', class_=['Tu-h3n Th3ty2'] )

    for n in name:
        title = n.find_all("span", class_=['Tu-h3tit'])
        noun.append(title[0].text)

    for i in info:
        miryang = i.find_all("span", class_=['TuTxtsp'])
    
        category.append(miryang[0].text)
        menu.append(miryang[1].text)  
        which.append(miryang[2].text)
    
    time.sleep(1)
    df = pd.DataFrame(
        {"업소명": noun,
         "카테고리": category,
         "주메뉴": menu, 
         "위치": which}
        
    )
df
    
```

```python
df.to_csv('밀양음식점전체.csv', index=False, encoding='euc-kr')
```
